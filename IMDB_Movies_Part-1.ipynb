```python
# %%
import pandas as pd 
import numpy as np
import seaborn as sns
```


# BASIC DATAFRAME

```python
# 
## uploaded file 
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
basics.head()
```

```python
basics['titleType'].value_counts()
```

```python
# %%
# replace missing values with NaN
basics.replace({'\\N':np.nan}, inplace=True)
```

```python
# %%
# get basic's dataframe info
basics.info()
```

```python
# 
# Check for missing values 
basics.isna().sum()
```

```python
###
# drop missing values from startYear, genres, runtimeMinutes

basics = basics.dropna(subset=['startYear', 'genres', 'runtimeMinutes', 'endYear'])
```

```python
## convert dytpe "startYear" to string
basics = basics.infer_objects().convert_dtypes()
```

```python
# check for missing values 
basics.isna().sum()
```

```python
#
#Keep only titleType_movie & Keep startYear 2000-2022

keepTs_df = basics[(basics['startYear']>='2000') & (basics['startYear']<='2022')]
keepTs_df
```

```python
movie_filter = basics['titleType']=='movie'
movie_filter
```

```python
#
new_df = keepTs_df
new_df
```

```python
new_df['titleType'].value_counts()
```

```python
# Exclude movies that are included in the documentary category. 
is_documentary = new_df['genres'].str.contains('Documentary', case=False)
df = new_df[~is_documentary]
```


# AKAS DATAFRAME / EDA 

```python
#
# Upload 2nd dataframe
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
df_akas = akas
df_akas.head()
```

```python
# replace missing values with NaN
df_akas.replace({'\\N':np.nan}, inplace=True)
```

```python
# US Filter 
df_filter_akas = df_akas[df_akas['region'] == 'US']
df_filter_akas
```

# RATINGS DATAFRAME

```python
#
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
df_ratings = ratings
df_ratings.head()
```

```python
#
# replace missing values with NaN
df_ratings.replace({'\\N':np.nan}, inplace=True)
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keeper = df_ratings['tconst'].isin(df_filter_akas['titleId'])
keeper
```

```python
# 
df_ratings_clean = df_ratings[keeper]
df_ratings_clean
```
